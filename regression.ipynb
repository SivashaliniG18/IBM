```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/50_Startups (1).csv')
```

```python
df.info()
```

```python
df['State'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.head()
```

```python
df['State']=le.fit_transform(df['State'])
df.head()
```

#split

```python
x=df.iloc[:,0:4]
```

```python
x
```

```python
y=df.iloc[:,4:5]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
regressor = Sequential()
regressor.add(Dense(4,activation='relu')) #input layer
regressor.add(Dense(12,activation='relu')) #hidden
regressor.add(Dense(8,activation='relu')) #hidden
regressor.add(Dense(9,activation='relu')) #hidden
regressor.add(Dense(1,activation='linear'))#ouput
```

```python
regressor.compile(optimizer='adam',loss='mse',metrics=['mse'])
```

Back propagation 

```python
regressor.fit(xtrain,ytrain,batch_size=10,epochs=300)
```

```python
ypred=regressor.predict(xtrain)
```
